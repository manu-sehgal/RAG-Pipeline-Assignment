```python
import csv
from bs4 import BeautifulSoup
import requests
import os
```

```python
import pandas as pd
```

```python
url = 'https://docs.nvidia.com/cuda/'
```

```python
f = open('NVIDIA.txt', 'a')
```

Root -> level 2 sublinks -> level 3 sublinks -> level 4 sublinks -> level 5 sublinks

```python
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
soup.prettify()
```

```python
print(soup)
```

## Scraping All text from the root url

```python
level1_txt = soup.body.get_text(' ', strip=True)    # scrape all copyable text from the page
f.write(level1_txt)
```

## Scraping all Links from the root url

```python
anchor_tags = soup.find_all("a")
print(anchor_tags)
links = [tag.get('href') for tag in anchor_tags]      # get all the web links present in the page
print(links)
```

```python
level2_webpages = set()  # not list, to filter any duplicates
```

```python
for link in links:
  elif link.startswith('http'):      # complete links present in the page
    level2_webpages.add(link)
  else:                              # incomplete links, we will prefix them with the current page's url so they form sublinks
    level2_webpages.add(url + link)
```

```python
level2_webpages
```

## Scraping all text from all level 2 sublinks

```python
level2_txt = ''
level3_webpages = set()
```

```python
count = 0
for link in level2_webpages:
  page = requests.get(link, allow_redirects=False)
  soup = BeautifulSoup(page.content, 'html.parser')
  if soup.body != None:
    txt = soup.body.get_text(' ', strip=True)
    level2_txt += ('' if txt in level2_txt else txt+'\n')  # checking if scraped text is already present in data
```

```python
f.write(level2_txt)
```

```python
len(level2_txt)
```

## Scraping all links from level 2 sublinks

```python
for url in level2_webpages:
  page = requests.get(url, allow_redirects=False)
  soup = BeautifulSoup(page.content, 'html.parser')
  anchor_tags = soup.find_all("a")
  links = [tag.get('href') for tag in anchor_tags]
  for link in links:
    if link != None and link.startswith('#') and link != '#':
      level3_webpages.add(url + link)
```
